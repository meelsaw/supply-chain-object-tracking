```python
import torch
from matplotlib import pyplot as plt
from PIL import Image
import os
```

```python

# Use your local YOLOv5 submodule
model = torch.hub.load('../yolov5', 'yolov5s', source='local')  # or yolov5n for nano model

# Confirm it's loaded
print(model)
```

```python
# Load test image
img_path = '../data/test_image_2.jpg'  # replace with your file name
img = Image.open(img_path)
img
```

```python
# Inference
results = model(img)

# Show results
results.print()     # Console summary
results.show()      # Display image with boxes
```
